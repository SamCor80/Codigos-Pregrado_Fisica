```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo .csv
archivo_csv = "ruta_del_archivo.csv"
datos = pd.read_csv(archivo_csv)

# Extraer las columnas en listas
time = datos['Time'].tolist()
illuminance = datos['Illuminance'].tolist()

# Función para agrupar y promediar datos cercanos
def procesar_datos(time, illuminance, umbral=10):
    promedios_illuminance = []
    tiempos_centrales = []

    grupo_illuminance = [illuminance[0]]
    grupo_time = [time[0]]

    for i in range(1, len(illuminance)):
        if abs(illuminance[i] - grupo_illuminance[-1]) <= umbral:
            grupo_illuminance.append(illuminance[i])
            grupo_time.append(time[i])
        else:
            # Calcular promedio y tiempo central
            promedios_illuminance.append(np.mean(grupo_illuminance))
            tiempos_centrales.append(np.mean(grupo_time))

            # Reiniciar los grupos
            grupo_illuminance = [illuminance[i]]
            grupo_time = [time[i]]

    # Para el último grupo
    if grupo_illuminance:
        promedios_illuminance.append(np.mean(grupo_illuminance))
        tiempos_centrales.append(np.mean(grupo_time))

    return tiempos_centrales, promedios_illuminance

# Procesar los datos
umbral_cercania = 10  # Define el umbral de cercanía
time_procesado, illuminance_procesado = procesar_datos(time, illuminance, umbral=umbral_cercania)

# Graficar los datos procesados
plt.figure(figsize=(10, 6))
plt.plot(time_procesado, illuminance_procesado, marker='o', linestyle='-', label="Datos procesados")
plt.xlabel("Time")
plt.ylabel("Illuminance")
plt.title("Datos procesados de Time e Illuminance")
plt.legend()
plt.grid()
plt.show()
```
