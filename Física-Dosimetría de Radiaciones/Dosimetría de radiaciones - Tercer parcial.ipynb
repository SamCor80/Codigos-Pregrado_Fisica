```python
# Paso 1: Montar Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Paso 2: Importar librerías necesarias
import os
import glob
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import zipfile

# Define la ruta de la carpeta de entrada en tu Drive y la carpeta de salida local en Colab
input_folder = '/content/drive/MyDrive/SPECT_Cts/scan1/ct'  # Reemplaza con tu ruta
output_folder = '/content/converted_images'
os.makedirs(output_folder, exist_ok=True)

# Paso 3: Buscar y procesar archivos .dmc
# Se usa glob para obtener todos los archivos con extensión .dmc
files = glob.glob(os.path.join(input_folder, '*.dcm'))
print(f"Se encontraron {len(files)} archivos .dcm.")

for file in files:
    try:
        # Se usa force=True por si la extensión no es la típica de DICOM (.dcm)
        ds = pydicom.dcmread(file, force=True)

        # Extraer la imagen; se asume que el archivo contiene un array de píxeles
        image = ds.pixel_array.astype(float)

        # Normalización de la imagen para el rango [0, 255]
        image = (np.maximum(image, 0) / image.max()) * 255.0
        image = np.uint8(image)

        # Construir el nombre del archivo de salida en formato PNG
        base = os.path.basename(file)
        name, _ = os.path.splitext(base)
        output_path = os.path.join(output_folder, name + '.png')

        # Guardar la imagen en escala de grises
        plt.imsave(output_path, image, cmap='gray')
        print(f"Convertido: {file} -> {output_path}")
    except Exception as e:
        print(f"Error procesando {file}: {e}")

# Paso 4: Comprimir la carpeta de imágenes convertidas en un archivo ZIP
zip_filename = '/content/converted_images.zip'
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            filepath = os.path.join(root, file)
            zipf.write(filepath, arcname=file)
print(f"\nImágenes convertidas y comprimidas en: {zip_filename}")
```

```python
# Rutas de las carpetas en tu Google Drive
ct_folder = "/content/drive/MyDrive/Tercer_parcial_dosimetria/scan1 - ct"  # Cambia con tu ruta real
spect_folder = "/content/drive/MyDrive/Tercer_parcial_dosimetria/scan1 - spect"  # Cambia con tu ruta real
output_folder = "/content/fused_images1"
os.makedirs(output_folder, exist_ok=True)

# Obtener listas de archivos ordenados
ct_files = sorted(glob.glob(os.path.join(ct_folder, "*.png")))
spect_files = sorted(glob.glob(os.path.join(spect_folder, "*.png")))[:130]  # Tomar solo las primeras 130 de SPECT

# Verificar la cantidad de imágenes
print(f"Se encontraron {len(ct_files)} imágenes CT y {len(spect_files)} imágenes SPECT (limitado a 130).")

# Procesar cada par de imágenes CT-SPECT
num_pairs = min(len(ct_files), len(spect_files))
print(f"Se procesarán {num_pairs} pares de imágenes.")

for i in range(num_pairs):
    ct_path = ct_files[i]
    spect_path = spect_files[i]

    # Cargar imágenes
    ct_image = cv2.imread(ct_path, cv2.IMREAD_GRAYSCALE)  # Imagen anatómica en escala de grises
    spect_image = cv2.imread(spect_path, cv2.IMREAD_GRAYSCALE)  # Imagen de dosis

    # Redimensionar SPECT a la forma de CT si es necesario
    if ct_image.shape != spect_image.shape:
        spect_image = cv2.resize(spect_image, (ct_image.shape[1], ct_image.shape[0]))

    # Normalizar ambas imágenes al rango [0, 255]
    ct_image = cv2.normalize(ct_image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    spect_image = cv2.normalize(spect_image, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    # **Solución al error: Convertir CT a 3 canales**
    ct_image_rgb = cv2.cvtColor(ct_image, cv2.COLOR_GRAY2BGR)

    # Aplicar colormap a la imagen SPECT (esto ya tiene 3 canales)
    spect_colormap = cv2.applyColorMap(spect_image, cv2.COLORMAP_JET)

    # Fusionar imágenes con transparencia (ahora ambas tienen 3 canales)
    alpha = 0.6  # Ajusta este valor para cambiar la intensidad de SPECT
    fused_image = cv2.addWeighted(ct_image_rgb, 1, spect_colormap, alpha, 0)

    # Guardar la imagen fusionada
    output_name = os.path.basename(ct_path).replace('.png', '_fusion.png')
    output_path = os.path.join(output_folder, output_name)
    cv2.imwrite(output_path, fused_image)
    print(f"Guardado: {output_name}")

# Paso 5: Comprimir las imágenes fusionadas en un ZIP
zip_filename = "/content/fused_images.zip"
with zipfile.ZipFile(zip_filename, "w") as zipf:
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            zipf.write(os.path.join(root, file), arcname=file)

print(f"\nImágenes fusionadas y comprimidas en: {zip_filename}")
```

EXTRACCIÓN DE TIEMPOS DE INYECCIÓN Y ADQUISICIONES

```python
import pydicom
import datetime
from google.colab import files

# Función para convertir formato DICOM (YYYYMMDD y hhmmss) a datetime
def dicom_datetime(date, time):
    try:
        if date != "Desconocido" and time != "Desconocido":
            date = date.strip()  # Eliminar espacios en blanco
            time_clean = time.split('.')[0].strip()  # Eliminar decimales extra y espacios
            return datetime.datetime.strptime(date + time_clean, "%Y%m%d%H%M%S")
    except Exception as e:
        print(f"Error al convertir el tiempo {time}: {e}")
        return None
    return None

# Subir archivo manualmente en Google Colab
uploaded_files = files.upload()

# Iterar sobre los archivos subidos
for filename in uploaded_files.keys():
    try:
        # Cargar archivo DICOM
        ds = pydicom.dcmread(filename, force=True)

        # Extraer la fecha y hora de adquisición
        acquisition_date = ds.get((0x0008, 0x0022), "").value if (0x0008, 0x0022) in ds else "Desconocido"
        acquisition_time = ds.get((0x0008, 0x0032), "").value if (0x0008, 0x0032) in ds else "Desconocido"

        # Buscar dentro de la secuencia (0054,0016) Radiopharmaceutical Information Sequence
        radionuclide_start_time = "Desconocido"
        radionuclide_start_date = "Desconocido"

        if hasattr(ds, "RadiopharmaceuticalInformationSequence"):
            seq = ds.RadiopharmaceuticalInformationSequence
            if len(seq) > 0:
                if (0x0018, 0x1072) in seq[0]:  # Hora de inyección
                    radionuclide_start_time = seq[0][(0x0018, 0x1072)].value
                if (0x0018, 0x1078) in seq[0]:  # Fecha de inyección
                    radionuclide_start_date = seq[0][(0x0018, 0x1078)].value[:8]  # Extraer solo YYYYMMDD

        # Verificar si los valores de inyección y adquisición están disponibles
        if radionuclide_start_time == "Desconocido" or radionuclide_start_date == "Desconocido":
            injection_datetime = None
        else:
            injection_datetime = dicom_datetime(radionuclide_start_date, radionuclide_start_time)

        if acquisition_date == "Desconocido" or acquisition_time == "Desconocido":
            acquisition_datetime = None
        else:
            acquisition_datetime = dicom_datetime(acquisition_date, acquisition_time)

        # Calcular diferencia en horas considerando la diferencia de días
        if injection_datetime and acquisition_datetime:
            time_difference = (acquisition_datetime - injection_datetime).total_seconds() / 3600
        else:
            time_difference = "No disponible"

        # Imprimir resultados
        print(f"Archivo: {filename}")
        print(f"Fecha de adquisición: {acquisition_date}")
        print(f"Hora de inyección: {radionuclide_start_time if radionuclide_start_time != 'Desconocido' else 'No disponible'}")
        print(f"Hora de adquisición: {acquisition_time if acquisition_time != 'Desconocido' else 'No disponible'}")
        print(f"Diferencia en horas: {time_difference} h")

    except Exception as e:
        print(f"Error procesando {filename}: {e}")
```

```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')

# Ruta de la carpeta donde están las imágenes DICOM en Google Drive
dicom_folder = "/content/drive/My Drive/Tercer_parcial_dosimetria/ROI"  # Cambia "TU_CARPETA" por el nombre real

# Lista de archivos DICOM en la carpeta
dicom_files = sorted([f for f in os.listdir(dicom_folder) if f.endswith('.dcm')])

# Coordenadas de la ROI (x_min:x_max, y_min:y_max)
x_min, x_max = 173, 188
y_min, y_max = 129, 142

# Lista para almacenar la intensidad media de cada imagen
intensidad_media = []

# Diccionario para almacenar los valores calculados
resultados = {}

# Procesar cada imagen DICOM
for idx, file in enumerate(dicom_files):
    filepath = os.path.join(dicom_folder, file)
    dicom_data = pydicom.dcmread(filepath)

    # Convertir a array de numpy (escala de grises)
    image_array = dicom_data.pixel_array.astype(np.float32)

    # Extraer la ROI
    roi = image_array[y_min:y_max, x_min:x_max]

    # Calcular estadísticas de la ROI
    min_intensity = np.min(roi)
    max_intensity = np.max(roi)
    mean_intensity = np.mean(roi)
    std_intensity = np.std(roi)

    # Guardar la intensidad media
    intensidad_media.append(mean_intensity)

    # Guardar resultados en el diccionario
    resultados[file] = {
        "Min": min_intensity,
        "Max": max_intensity,
        "Media": mean_intensity,
        "Desviación Estándar": std_intensity
    }

# Mostrar los resultados en una tabla
df_resultados = pd.DataFrame(resultados).T
display(df_resultados)  # Muestra los datos en Colab

# Eje x (tiempo)
t = [3.74, 27.72, 103.09, 123.98]

# Graficar la intensidad media vs. tiempo
plt.figure(figsize=(8, 5))
plt.plot(t, intensidad_media, 'ob', label='Intensidad media')

# Ajustar tamaños de letra
plt.xlabel('Tiempo (h)', fontsize=14)
plt.ylabel('Intensidad media', fontsize=14)
plt.title('Evolución de Intensidad Media en la ROI', fontsize=15)
plt.legend(fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(True)
plt.show()
```

```python
import pydicom
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from scipy.optimize import curve_fit
from scipy.stats import linregress
from google.colab import drive

# Montar Google Drive
drive.mount('/content/drive')

# Ruta de la carpeta donde están las imágenes DICOM en Google Drive
dicom_folder = "/content/drive/My Drive/Tercer_parcial_dosimetria/ROI"  # Cambia "TU_CARPETA" por el nombre real

# Lista de archivos DICOM en la carpeta
dicom_files = sorted([f for f in os.listdir(dicom_folder) if f.endswith('.dcm')])

# Coordenadas de la ROI (x_min:x_max, y_min:y_max)
x_min, x_max = 173, 188
y_min, y_max = 129, 142

# Listas para almacenar datos
intensidad_media = []
desviaciones_std = []

# Diccionario para almacenar los valores calculados
resultados = {}

# Procesar cada imagen DICOM
for idx, file in enumerate(dicom_files):
    filepath = os.path.join(dicom_folder, file)
    dicom_data = pydicom.dcmread(filepath)

    # Convertir a array de numpy (escala de grises)
    image_array = dicom_data.pixel_array.astype(np.float32)

    # Extraer la ROI
    roi = image_array[y_min:y_max, x_min:x_max]

    # Calcular estadísticas de la ROI
    min_intensity = np.min(roi)
    max_intensity = np.max(roi)
    mean_intensity = np.mean(roi)
    std_intensity = np.std(roi)

    # Guardar valores en listas
    intensidad_media.append(mean_intensity)
    desviaciones_std.append(std_intensity)

    # Guardar resultados en el diccionario
    resultados[file] = {
        "Min": min_intensity,
        "Max": max_intensity,
        "Media": mean_intensity,
        "Desviación Estándar": std_intensity
    }

# Mostrar los resultados en una tabla
df_resultados = pd.DataFrame(resultados).T
display(df_resultados)  # Muestra los datos en Colab

# Eje x (tiempo)
t = np.array([3.74, 27.72, 103.09, 123.98])
intensidad_media = np.array(intensidad_media)
desviaciones_std = np.array(desviaciones_std)

# Definir la función de ajuste exponencial
def exp_fit(x, A, B):
    return A * np.exp(B * x)

# Ajuste exponencial
popt, pcov = curve_fit(exp_fit, t, intensidad_media, p0=(max(intensidad_media), -0.01))

# Generar datos para la curva ajustada
t_fit = np.linspace(min(t), max(t), 100)
y_fit = exp_fit(t_fit, *popt)

# Graficar la intensidad media vs. tiempo con ajuste exponencial y barras de error
plt.figure(figsize=(8, 5))
plt.errorbar(t, intensidad_media, yerr=desviaciones_std, fmt='ob', label='Datos')
plt.plot(t_fit, y_fit, 'r--', label=f'Ajuste exp: $I = {popt[0]:.2f} e^{{{popt[1]:.4f} t}}$')

# Ajustar tamaños de letra
plt.xlabel('Tiempo (h)', fontsize=14)
plt.ylabel('Intensidad media', fontsize=14)
plt.title('Evolución de Intensidad Media en la ROI', fontsize=15)
plt.legend(fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()

# ----------------------
# **Escala Semilogarítmica y Ajuste Lineal**
# ----------------------

# Transformar los valores a logaritmo natural
log_intensidad = np.log(intensidad_media)

# Ajuste lineal (y = mx + b en escala log)
slope, intercept, r_value, p_value, std_err = linregress(t, log_intensidad)

# Generar datos para la recta ajustada
log_fit = slope * t_fit + intercept

# Graficar en escala semilogarítmica con ajuste lineal
plt.figure(figsize=(8, 5))
plt.errorbar(t, log_intensidad, yerr=desviaciones_std / intensidad_media, fmt='ob', label='Datos')
plt.plot(t_fit, log_fit, 'r--', label=f'Ajuste lin: $\lambda = {slope:.4f}$, $R^2 = {r_value**2:.4f}$')

# Ajustar tamaños de letra
plt.xlabel('Tiempo (h)', fontsize=14)
plt.ylabel('ln(Intensidad)', fontsize=14)
plt.title('Escala Semilogarítmica: Ajuste Lineal', fontsize=15)
plt.legend(fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.show()

# Mostrar la pendiente y el coeficiente de determinación R^2
print(f"Pendiente de la recta ajustada: {slope:.4f}")
print(f"Coeficiente de determinación (R^2): {r_value**2:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los parámetros de la tabla
t = np.linspace(0, 130, 200)  # Tiempo desde 0 hasta 130 horas

# Parámetros para cada modelo
params = {
    "Físico": {"lambda": 0.0063, "A0": 7210.43},
    "Lineal": {"lambda": 0.0161, "A0": 5721.84},
    "Exponencial": {"lambda": 0.0143, "A0": 5415.30},
}

# Colores para cada modelo
colors = {"Físico": "g", "Lineal": "b", "Exponencial": "r"}

# Graficar la actividad para cada modelo
plt.figure(figsize=(8, 5))

for model, data in params.items():
    A_t = data["A0"] * np.exp(-data["lambda"] * t)
    plt.plot(t, A_t, linestyle="-.", color=colors[model], label=f"{model}: $A_0$={data['A0']}, $\lambda$={data['lambda']:.4f}")

# Ajustes de la gráfica
plt.xlabel('Tiempo (h)', fontsize=14)
plt.ylabel('Actividad (MBq)', fontsize=14)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)

# Mostrar la gráfica
plt.show()
```
