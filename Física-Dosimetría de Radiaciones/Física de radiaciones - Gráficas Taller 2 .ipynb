**2. Calcule la energía del fotón dispersado por Compton a θ=0°, 45°, 90°, y 180°, para energías de 50 keV, 500 keV y 5 MeV. Grafique sus resultados**

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros iniciales:

E0 = [0.05, 0.5, 5]
theta = [0, np.pi/4, np.pi/2, np.pi]
mc2 = 0.511

# Energía del fotón dispersado:

def E_compton(E0, theta):
    return E0/(1+E0*(1-np.cos(theta))/mc2)

E = np.zeros((len(E0), len(theta)))
for i in range(len(E0)):
  for j in range(len(theta)):
    E[i,j] = E_compton(E0[i], theta[j])
plt.figure()
plt.loglog(E0, E[:,0], 'o', label=r'$\theta=0^\circ$', markersize=5)
plt.loglog(E0, E[:,1], 's', label=r'$\theta=45^\circ$', markersize=5)
plt.loglog(E0, E[:,2], 'd', label=r'$\theta=90^\circ$', markersize=6)
plt.loglog(E0, E[:,3], '^', label=r'$\theta=180^\circ$', markersize=7)
plt.xlabel('$E_0$ (MeV)')
plt.ylabel('Energía final (MeV)')
plt.title('Energía del fotón dispersado por Compton (log-log)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.xticks(E0, [str(val) for val in E0])
plt.yticks(E0, [str(val) for val in E0])
plt.show()
```

**3. ¿Cuáles son los ángulos de dispersión de los electrones correspondientes a las energías del anterior ejercicio? Grafique sus resultados**

```python
import numpy as np
import matplotlib.pyplot as plt

E0 = [0.05, 0.5, 5]
theta = [0, np.pi/4, np.pi/2, np.pi]
mc2 = 0.511

def Angulo_Electron(E0, theta):
  CotPhi = (1+E0/mc2)*np.tan(theta/2)
  return np.arctan(1/CotPhi)

Angulo = np.zeros((len(E0), len(theta)))
for i in range(len(E0)):
  for j in range(len(theta)):
    if theta[j] == 0:
      Angulo[i, j] = np.pi/2
    else:
      Angulo[i, j] = Angulo_Electron(E0[i], theta[j])

# Graficas:
plt.figure(figsize=(8, 6))
plt.semilogx(E0, Angulo[:,0], 'o', markersize=7, label=r'$\theta_\gamma=0^\circ$')
plt.semilogx(E0, Angulo[:,1], 's', markersize=7, label=r'$\theta_\gamma=45^\circ$')
plt.semilogx(E0, Angulo[:,2], 'd', markersize=7, label=r'$\theta_\gamma=90^\circ$')
plt.semilogx(E0, Angulo[:,3], '^', markersize=7, label=r'$\theta_\gamma=180^\circ$')
plt.xlabel('Energía incidente $E_0$ (MeV)', fontsize=12)
plt.ylabel('Ángulo del electrón $\phi$ (radianes)', fontsize=12)
plt.title('Ángulo de dispersión del electrón en función de $E_0$', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='best', fontsize=10)
plt.xticks(E0, [str(e) for e in E0])
plt.tight_layout()
plt.show()
```
