```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

filename_p = 'Rango_protones.txt'
filename_a = 'Rango_alfas.txt'

# Leer y procesar los datos
energy_p = []  # Energa (MeV)
csda_p = []    # Rango CSDA (g/cm)
projected_p = []  # Rango Proyectado (g/cm)

with open(filename_p, 'r') as file:
    lines = file.readlines()
    start_reading = False

    for line in lines:
        # Buscar el inicio de los datos numricos
        if line.startswith('1.000E+00'):
            start_reading = True

        if start_reading:
            data = line.split()
            if len(data) == 3:
                try:
                    energy_p.append(float(data[0]))
                    csda_p.append(float(data[1]))
                    projected_p.append(float(data[2]))
                except ValueError:
                    continue

energy_a = []  # Columna 1: Energa (MeV)
csda_a = []    # Columna 2: Rango CSDA (g/cm)
projected_a = []  # Columna 3: Rango Proyectado (g/cm)

with open(filename_a, 'r') as file:
    lines = file.readlines()
    start_reading = False

    for line in lines:
        # Buscar el inicio de los datos numricos
        if line.startswith('1.000E+00'):
            start_reading = True

        if start_reading:
            data = line.split()
            if len(data) == 3:
                try:
                    energy_a.append(float(data[0]))
                    csda_a.append(float(data[1]))
                    projected_a.append(float(data[2]))
                except ValueError:
                    continue

# Crear la grfica
plt.figure(figsize=(12, 7))
plt.loglog(energy_p, csda_p, 'bo-', label='Rango CSDA - Protones', linewidth=3, markersize=6)
plt.loglog(energy_p, projected_p, 'rs', label='Rango Proyectado - Protones', linewidth=2, markersize=6)
plt.loglog(energy_a, csda_a, 'go-', label='Rango CSDA - Alfas', linewidth=3, markersize=6)
plt.loglog(energy_a, projected_a, 'ms', label='Rango Proyectado - Alfas', linewidth=2, markersize=6)

# Configuracin de la grfica
plt.title('Rangos de Protones y Alfas en Agua', fontsize=14)
plt.xlabel('Energa Cintica (MeV)', fontsize=12)
plt.ylabel('Rango (g/cm)', fontsize=12)
plt.grid(True, which='both', linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
