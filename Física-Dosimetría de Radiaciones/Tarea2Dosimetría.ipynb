```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parmetros del espectro
E_peak = 662  # keV, energa del pico
sigma_Fano1 = 10  # keV, desviacin estndar con Fano = 1
sigma_Fano6 = np.sqrt(6.04) * sigma_Fano1  # Aumenta la desviacin con Fano = 6.04

# Rango de energas
E = np.linspace(620, 700, 1000)

# Distribuciones gaussianas para los dos casos
gauss_Fano1 = norm.pdf(E, E_peak, sigma_Fano1)
gauss_Fano6 = norm.pdf(E, E_peak, sigma_Fano6)

# Normalizar para comparar mejor
gauss_Fano1 /= max(gauss_Fano1)
gauss_Fano6 /= max(gauss_Fano6)

# Graficar los espectros
plt.figure(figsize=(8, 5))
plt.plot(E, gauss_Fano1, label="Fano = 1 (Menor dispersin)", linewidth=2)
plt.plot(E, gauss_Fano6, label="Fano = 6.04 (Mayor dispersin)", linewidth=2, linestyle="dashed")

# Etiquetas y leyenda
plt.xlabel("Energa (keV)", size=12)
plt.ylabel("Intensidad Normalizada", size=12)
plt.title("Efecto del Factor de Fano en la Resolucin Energtica", size=15)
plt.legend()
plt.grid()

# Mostrar la grfica
plt.show()
```

```python
# Reintentar la generacin de la grfica tras el reinicio

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definir energas de entrada (energas incidentes)
E_incidentes = [200, 400, 600, 800, 1000]  # keV

# Definir una resolucin tpica del detector (desviacin estndar)
sigma = 30  # keV (resolucin del detector)

# Rango de energas de salida (energas medidas por el detector)
E_medidas = np.linspace(100, 1100, 1000)

# Generar curvas gaussianas para cada energa incidente
plt.figure(figsize=(8, 5))
for E in E_incidentes:
    respuesta = norm.pdf(E_medidas, E, sigma)
    respuesta /= max(respuesta)  # Normalizar para mejor visualizacin
    plt.plot(E_medidas, respuesta, label=f"E_inc = {E} keV")

# Etiquetas y leyenda
plt.xlabel("Energa medida (keV)")
plt.ylabel("Probabilidad Normalizada")
plt.title("Ejemplo de Funcin Respuesta de un Detector")
plt.legend()
plt.grid()

# Mostrar la grfica
plt.show()
```
