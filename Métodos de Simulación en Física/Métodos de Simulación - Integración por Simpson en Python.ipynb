Librerías y variables globales:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy as sci
```

Funciones globales:

```python
def f(t):
  return np.cos(t)
```

```python
def IntegralPorSimpson(a,b,n):
  n*=2  # Aseguramos que n sea par
  h=(b-a)/n
  suma=0
  for i in range(n+1):
    x=a+i*h
    if i==0 or i==n:
      suma+=f(x)
    elif i%2==0:
      suma+=2*f(x)
    else:
      suma+=4*f(x)
  return suma*h/3
```

Programa principal:

```python
a=0
b=np.pi/3
n=50

Result=IntegralPorSimpson(a,b,n)
print("La integral de la función por Simpson es:",Result)
```

Dado que podemos definir a las funciones de Bessel por medio de una integral, podemos utilizar el método de Simpson para calcular y graficar estas funciones:

```python
def f(t,alpha,x):
  return np.cos(alpha*t-x*np.sin(t))
```

```python
def IntegralPorSimpsonBessel(a,b,n,alpha,x):
  n*=2
  h=(b-a)/n
  suma=0
  for i in range(n+1):
    t=a+i*h
    if i==0 or i==n:
      suma+=f(t,alpha,x)
    elif i%2==0:
      suma+=2*f(t,alpha,x)
    else:
      suma+=4*f(t,alpha,x)
  return suma*h/3
```

```python
def Bessel(alpha,x):
  return IntegralPorSimpsonBessel(0,np.pi,100,alpha,x)
```

```python
# Programa principal:

x0=0
xf=10
nsteps=100
dx=(xf-x0)/nsteps

alpha=0
xdata=np.zeros(nsteps)
ydata=np.zeros(nsteps)
for i in range(nsteps):
  x=x0+i*dx
  xdata[i]=x
  ydata[i]=Bessel(alpha,x)

plt.plot(xdata,ydata,'--r')
plt.xlabel("$x$",size=15)
plt.ylabel("$J_{0}(x)$",size=15)
plt.title("Función de Bessel de orden 0",size=16)
plt.show()
```
