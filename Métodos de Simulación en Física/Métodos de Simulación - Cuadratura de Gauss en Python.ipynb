Funciones globales:

```python
def f(x):  # Defino la función que deseo integrar
  return np.sin(x)
```

Haremos un ejemplo de programa para calcular la integral de una función mediante cuadratura de Gauss, utilizando polinomios de Legendre:

```python
def CuadraturaDeGauss(a,b):
  Medioh=(b-a)/2
  xprom=(a+b)/2
  orden_n=3
  # Ceros del polinomio de Legendre de orden 3, p_3(x)
  xi=np.zeros(orden_n)
  xi[0]=-np.sqrt(3/5)
  xi[1]=0
  xi[2]=np.sqrt(3/5)
  # Pesos de esos mismos puntos, calculados con la interpolación de Lagrange
  w=np.zeros(orden_n)
  w[0]=5/9
  w[1]=8/9
  w[2]=5/9
  # Aproximamos la integral por la sumatoria:
  suma=0
  for i in range(orden_n):
    # Primero debemos trasladar los puntos xi, desde el intervalo [-1,1] hasta el intervalo [a,b]
    x=Medioh*xi[i]+xprom # Revisar este paso!!
    suma+=w[i]*f(x)
  return Medioh*suma
```

El resultado anterior es la definición de la cuadratura de Gauss en un intervalo [a,b]. Ahora, sólo nos falta definir la cuadratura de Gauss en los n intervalos totales:

```python
def CuadraturaDeGaussEn_n_Intervalos(a,b,nsteps):
  h=(b-a)/nsteps
  suma=0
  for i in range(nsteps):
    ak=a+i*h
    suma+=CuadraturaDeGauss(ak,ak+h)
  return suma
```
