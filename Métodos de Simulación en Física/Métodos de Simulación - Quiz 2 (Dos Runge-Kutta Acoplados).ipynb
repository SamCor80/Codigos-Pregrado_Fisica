```python
import numpy as np
import matplotlib.pyplot as plt

# Algunos parámetros y constantes:

T = np.pi
omega = 2*np.pi/T

# Definimos las ecuaciones acopladas:

def f(t,x,y):
  return y

def g(t,x,y):
  return -omega**2 * x

# Condiciones iniciales y MRK4:

x = 0
y = 1
t = 0
X = [x]
Y = [y]
T = [t]

dt = 0.1
while t<=10:
    dx1 = f(t,x,y)*dt
    dy1 = g(t,x,y)*dt
    dx2 = f(t+0.5*dt,x+0.5*dx1,y+0.5*dy1)*dt
    dy2 = g(t+0.5*dt,x+0.5*dx1,y+0.5*dy1)*dt
    dx3 = f(t+0.5*dt,x+0.5*dx2,y+0.5*dy2)*dt
    dy3 = g(t+0.5*dt,x+0.5*dx2,y+0.5*dy2)*dt
    dx4 = f(t+dt,x+dx3,y+dy3)*dt
    dy4 = g(t+dt,x+dx3,y+dy3)*dt
    x += (dx1+2*dx2+2*dx3+dx4)/6
    y += (dy1+2*dy2+2*dy3+dy4)/6
    t += dt
    X.append(x)
    Y.append(y)
    T.append(t)
```

```python
plt.plot(T,X,'--r')
plt.xlabel('$t$',size=15)
plt.ylabel('$x(t)$',size =15)
plt.title('Solución de EDO de segundo orden',size=16)
plt.show()
```
