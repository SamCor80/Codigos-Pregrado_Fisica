```python
# Librerías y variables globales (constantes):

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import scipy as sci

ERR = 1e-7
```

```python
# Funciones globales:

def f(x):
  return np.sin(x)/x
```

```python
def CerosPorBiseccion(a,b):
  fa=f(a)
  while b-a>ERR:
    m=(b+a)/2
    fm=f(m)
    if fa*fm>0:
      a=m
      fa=fm
    else:
      b=m
  return (a+b)/2
```

```python
# Programa principal:

a=2
b=4

print("El cero por bisección de la función es:",CerosPorBiseccion(a,b))
```
