```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Se sube el archivo .csv:

```python
reflectancia = pd.read_csv('Reflectancia.csv',sep=";",decimal=",",skiprows=2,names=['Lambda','R','Rt','FR','FRov0.5','hnu'])
```

```python
xmin=350
xmax=2500
ymin=0
ymax=100

plt.plot(reflectancia["Lambda"],reflectancia['R'], "-g")
plt.axis((xmin, xmax, ymin, ymax))
plt.xlabel("Longitud de onda (nm)",size=12)
plt.ylabel("% Reflectancia",size=12)
plt.savefig('Reflectancia_vs_longitud.jpg')
```
