```python
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram

# Crear un circuito cuántico con 3 qubits (2 para los estados y 1 auxiliar)
qc = QuantumCircuit(3, 1)

# Inicializar qubits |ψ⟩ y |φ⟩ aquí (ejemplo: ambos en estado |0⟩)
# Si quieres inicializar en otros estados, usa qc.initialize() o qc.u3().

# Aplicar la puerta Hadamard al qubit auxiliar
qc.h(2)

# Aplicar las puertas CNOT
qc.cx(0, 2)
qc.cx(1, 2)

# Aplicar otra puerta Hadamard al qubit auxiliar
qc.h(2)

# Medir el qubit auxiliar
qc.measure(2, 0)

# Ejecutar el circuito en el simulador
simulator = Aer.get_backend('qasm_simulator')
result = simulator.run(qc, shots=1024).result()

# Obtener y mostrar los resultados
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```
