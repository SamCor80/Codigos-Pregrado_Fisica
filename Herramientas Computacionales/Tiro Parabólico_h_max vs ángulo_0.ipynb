```python
import matplotlib.pyplot as plt
import numpy as np

#Definimos la función que nos permite calcular la altura máxima

g = 9.8 #Aceleración de la gravedad (en m/s^2)

def h_max(conditions):
    v_i = conditions['v0']
    theta_i = conditions['theta0']
    ymax = ((v_i ** 2)*(np.sin(theta_i) ** 2))/(2 * g)
    return ymax
```

```python
#Definimos las condiciones iniciales:

v0 = 2.34
thetas = np.linspace(0 , np.pi/2 , 20)
```

```python
#Creamos un bucle que nos permita calcular los valores de altura máxima en función del ángulo de lanzamiento, utilizando la función previamente definida:

mh_values = [] #Lista de valores de altura máxima
for i in thetas:
    initial_cond = {'v0' : v0,
                    'theta0' : i}
    mh_values.append(h_max(initial_cond))
```

```python
#Finalmente, realizamos la gráfica que nos muestre cómo cambia la altura máxima conforme varía el ángulo inicial:

plt.plot(thetas,mh_values, "-r", marker = ".")
plt.xlabel('Ángulo de lanzamiento (radianes)')
plt.ylabel('Altura máxima (metros)')
plt.title('Altura máxima en función del ángulo')
plt.grid()
plt.savefig('altura_max.pdf')
```
