## Caminante aleatorio en 2D:

En este caso, definimos un vector de posición *r=(x,y)* que dependa de un ángulo (entre *0* y *2π*) escogido de manera aleatoria:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 0
y = 0
X = [x]
Y = [y]
for t in range (10000):
  theta = 2*np.pi*np.random.rand()  # Se escoge un número aleatorio entre 0 y 1
  x += np.cos(theta)
  y += np.sin(theta)
  X.append(x)
  Y.append(y)
plt.plot(X,Y,'-g', linewidth=0.8)
plt.title('Caminante aleatorio en 2D',size = 20)
plt.xlabel('$x$',size = 15)
plt.ylabel('$y$',size = 15)
plt.show()
```

Este modelo, por ejemplo, describe el comportamiento de un grano de polen sometido a movimiento browniano.

```python
for i in range(50):  # Número de caminantes aleatoriosk
  x = 0
  y = 0
  X = [x]
  Y = [y]
  for t in range (1000):
    theta = 2*np.pi*np.random.rand()  # Se escoge un número aleatorio entre 0 y 1
    x += np.cos(theta)
    y += np.sin(theta)
    X.append(x)
    Y.append(y)
  plt.plot(X,Y,linewidth=0.8)
plt.title('Caminantes aleatorios en 2D',size = 20)
plt.xlabel('$x$',size = 15)
plt.ylabel('$y$',size = 15)
plt.show()
```

Si miramos las distribuciones de probabilidad de todos estos caminantes aleatorios, deberían estar descritas por una campana de Gauss en 3D

## Distribución de Pareto:
Está dada de la siguiente forma: \\
*ρ(x) = cx^(-α)* \\
Este tipo de distribuciones son sumamente importantes en el estudio de sistemas complejos.

```python
np.random.pareto(1.5)  # alpha = 1.5, toma un valor aleatorio de x, siguiendo la distribución de probabilidad de Pareto
```

# Cálculo fraccionario:

## Información no local:

En la física y en muchas ramas de las ciencias, no basta con tener sólo la información local del sistema para lograr describirlo correctamente. Tenemos procesos que tienen "memoria" y cuya evolución también depende del pasado o de información previa. La información no local es común en sistemas complejos, pues el fenómeno de emergencia puede verse, en cierta medida, condicionado por esta información previa (no local).

## Derivadas fraccionarias:
Las derivadas fraccionarias entran a jugar un rol importante dentro del estudio de la no localidad de los sistemas (de forma similar a como, p.e., las derivadas de orden natural nos sirven para describir la localidad de los fenómenos). También se conocen como derivadas de orden real (orden *α >= 0*).

# Vuelos de Lévy:


Los *vuelos de Lévy* están basados en el movimiento browniano en 2D. Sin embargo, en este caso, el tamaño del desplazamiento estará dado por una distribución de Pareto: \\
*P(Δl)∝Δl**β* \\
Es un concepto muy útil en varios campos de la ciencia, como la ecología (modelar el desplazamiento de algún animal dentro de un ecosistema, pues este tipo de dinámicas es útil para optimizar la búsqueda de recursos en la naturaleza).

```python
x = 0
y = 0
X = [x]
Y = [y]
for i in range(1000):
  theta = 2*np.pi*np.random.rand()
  r = np.random.pareto(2)
  x += r*np.cos(theta)
  y += r*np.sin(theta)
  X.append(x)
  Y.append(y)

plt.plot(X,Y,'k-',linewidth=0.8)
plt.title('Vuelos de Lévy',size = 20)
plt.xlabel('$x$',size = 15)
plt.ylabel('$y$',size = 15)
plt.show()
```
