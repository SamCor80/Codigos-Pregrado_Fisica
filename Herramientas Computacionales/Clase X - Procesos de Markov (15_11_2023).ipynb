```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
Data = pd.read_csv('nvda.us.txt')
```

```python
Data.head()
```

```python
Data.tail()
```

```python
plt.plot(Data['<OPEN>'])
plt.show()
```

```python
X = np.array(Data['<OPEN>'])
```

```python
Delta = (X[1::]-X[:-1])/X[:-1]
```

```python
plt.plot(Delta)
plt.ylabel(r'$\Delta$', size=20)
plt.show()
```

```python
freq,base = np.histogram(Delta,bins=100)
```

```python
len(freq)
```

```python
len(base)
```

```python
plt.plot(base[1::],freq,'m.')
plt.show()
```

```python
3*np.std(Delta)
```

```python
freq,base = np.histogram(np.abs(Delta),bins=100,density=True)
```

```python
plt.plot(base[1::],freq,'r.')
plt.show()
```

```python
plt.loglog(base[1::],freq,'r.')
plt.show()
```

```python
freq,base = np.histogram(np.abs(Delta),bins=10**np.linspace(-3,0,101),density=True)
```

```python
plt.loglog(base[1::],freq,'r.')
plt.show()
```
