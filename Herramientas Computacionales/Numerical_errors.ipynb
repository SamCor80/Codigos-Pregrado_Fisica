# Ecuación cuadrática:
Definimos la función que nos indique gráficamente la diferencia relativa entre ambos tipos de soluciones, fijando *a=1.0* y *b=1.0*:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def relative_error(a,b,N):
    error_1 = np.zeros(N)
    error_2 = np.zeros(N)
    c = np.zeros(N)
    for i in range(1,N+1):
        c[i-1] = 10 ** (-i)
        x1 = (- b + np.sqrt(np.square(b) - 4*a*c[i-1]))/(2*a)
        x2 = (- b - np.sqrt(np.square(b) - 4*a*c[i-1]))/(2*a)
        y1 = (-2*c[i-1])/(b + np.sqrt(np.square(b) - 4*a*c[i-1]))
        y2 = (-2*c[i-1])/(b - np.sqrt(np.square(b) - 4*a*c[i-1]))
        error_1[i-1] = np.abs(x1-y1)/np.abs(x1)
        error_2[i-1] = np.abs(x2-y2)/np.abs(x2)
    plt.plot(c,error_1,'ob',label = "Error entre x1 y x1'")
    plt.plot(c,error_2,'*y',label = "Error entre x2 y x2'")
    plt.xlabel('Valores de c')
    plt.ylabel('Error relativo')
    plt.title('Error relativo de las raíces')
    plt.semilogy()
    plt.semilogx()
    plt.legend()
    plt.grid()
    plt.show()


relative_error(1,1,16)
```

Como podemos observar, la diferencia relativa entre los valores de las raíces es mayor conforme tenemos valores de *c* más pequeños.
En este caso, *b^2>>ac* y *b>0*, por lo que computacionalmente es más conveniente utilizar *x1'* y *x2*, para que no se nos cancele ninguna raíz y poder obtener mayor exactitud en las soluciones.

# Arreglando una expresión matemática:
Si operamos algebráicamente, podemos darnos cuenta que podemos reescribir la función original en una forma equivalente de la siguiente forma:
*f(x) = 1 - sin(x) = 2sin^2(pi/4-x/2)*. Tenemos entonces:

```python
import numpy as np
import matplotlib.pyplot as plt

def function(c,N):
  f_original = np.zeros(2*N+1)
  f_eq = np.zeros(2*N+1)
  c_values = [c - 10**(-i) for i in range(1, N+1)] + [c] + [c + 10**(-i) for i in range(N, 0,-1)]
  for j in range(0,2*N+1):
        f_original[j] = 1 - np.sin(c_values[j])
        f_eq[j] = 2 * np.square(np.sin(np.pi/4-c_values[j]/2))
  plt.plot(c_values, f_original, '*r', label='f(x) = 1 - sin(x)')
  plt.plot(c_values, f_eq, '-.b', label='f(x) = 2*sin^2(pi/4 - x/2)')
  plt.xlabel('x')
  plt.ylabel('f(x)')

  plt.title('Comparación de las funciones')
  plt.legend()
  plt.grid(True)
  plt.show()

function(np.pi/2,10)

```

Si graficamos para los valores de *x* muy cercanos a *pi/2*, podremos observar que, mientras la función original cae relativamente rápido al cero, la función corregida nos da valores muy pequeños pero no nulos.
