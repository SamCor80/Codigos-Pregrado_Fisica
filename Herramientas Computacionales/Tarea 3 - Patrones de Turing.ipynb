```python
# Nombre: Samuel Cortes Salazar

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros:

a = 0.00028
b = 0.005
tau = 0.1
k = -0.005
h = 0.02
delta_t = 0.001
t_max = 10
T = np.arange(0,t_max+delta_t,delta_t)  # 0 <= t <= 10
X = np.arange(0,1+h,h)  # 0 <= x <=1
Y = np.arange(0,1+h,h)  # 0 <= y <=1

A,B,C = np.meshgrid(X,T,Y)
Psi = np.ones(A.shape)
Phi = np.ones(B.shape)
```

```python
#  Condiciones iniciales:

for i in range(0,len(X)):
  for j in range(0,len(Y)):
    Psi[0,i,j]=np.random.rand()
    Phi[0,i,j]=np.random.rand()
```

```python
for t in range(0,len(T)-1):
  for i in range(0,len(X)-1):
    for j in range(0,len(Y)-1):
      Lap_Phi = (Phi[t,i+1,j] + Phi[t,i-1,j] + Phi[t,i,j+1] + Phi[t,i,j-1] - 4*Phi[t,i,j])/h**2
      Lap_Psi = (Psi[t,i+1,j] + Psi[t,i-1,j] + Psi[t,i,j+1] + Psi[t,i,j-1] - 4*Psi[t,i,j])/h**2
      Phi[t+1,i,j] = Phi[t,i,j] + delta_t*(a*Lap_Phi + Phi[t,i,j] - Phi[t,i,j]**3 - Psi[t,i,j] + k)
      Psi[t+1,i,j] = Psi[t,i,j] + delta_t*(b*Lap_Psi + Phi[t,i,j] - Psi[t,i,j])/tau
      Phi[t,0,:] = Phi[t,1,:]
      Phi[t,-1,:] = Phi[t,-2,:]
      Psi[t,0,:] = Psi[t,1,:]
      Psi[t,-1,:] = Psi[t,-2,:]
      Phi[t,:,0] = Phi[t,:,1]
      Phi[t,:,-1] = Phi[t,:,-2]
      Psi[t,:,0] = Psi[t,:,1]
      Psi[t,:,-1] = Psi[t,:,-2]

tt = [0,2000,4000,6000,8000,10000]
for i in tt:
  plt.imshow(Phi[i,:,:],extent=(min(X),max(X),min(Y),max(Y)),interpolation='bilinear')
  plt.title(f'Solución $\phi$ en $t$={i/1000}s')
  plt.ylabel(r'$y$',size=20)
  plt.xlabel(r'$x$',size=20)
  plt.colorbar()
  plt.show()
  plt.imshow(Psi[0,:,:],extent=(min(X),max(X),min(Y),max(Y)),interpolation='bilinear')
  plt.title(f'Solución $\psi$ en $t$={i/1000}s')
  plt.ylabel(r'$y$',size=20)
  plt.xlabel(r'$x$',size=20)
  plt.colorbar()
  plt.show()
```
