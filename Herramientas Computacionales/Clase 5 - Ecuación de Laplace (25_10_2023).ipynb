# Forma discreta de la Ecuación de Laplace:

**Forma discreta de la segunda derivada:** Podemos comparar la información de la solución *f(x)* en los alrededores de *x* de la siguiente manera: \\
*d^2/dx^2(f(x)) ≈ [f(x+Δx)+f(x-Δx)-2f(x)]/(Δx)^2* \\
**Forma discreta de la primera derivada:** Está dada mediante la definición de la derivada:  \\
*d/dx(f(x)) ≈ [f(x+Δx)-f(x)]/(Δx),* \\
siempre que *Δx* sea lo suficientemente pequeño. *(Demostrado)*


**Laplaciano en 2D:** Se tiene, en coordenadas cartesianas, que: \\
*∇^2 = ∂^2/∂x^2 + ∂^2/∂y^2* \\
Luego: \\
*∇^2(Ψ(x,y)) ≈ [Ψ(x+Δx,y)+Ψ(x-Δx,y)-2Ψ(x,y)]/(Δx)^2 + [Ψ(x,y+Δy)+Ψ(x,y-Δy)-2Ψ(x,y)]/(Δy)^2* \\
Asumiendo que *Δx = Δy = Δl*, tenemos: \\
*∇^2(Ψ(x,y)) ≈ [Ψ(x+Δl,y)+Ψ(x-Δl,y)+Ψ(x,y+Δl)+Ψ(x,y-Δl)-4Ψ(x,y)]/(Δl)^2* \\




**Ecuación de Laplace en 2D:** A partir de la definición del laplaciano en su forma discreta, podemos definir la Ecuación de Laplace (*∇^2(Ψ(x,y) = 0*) como: \\
*[Ψ(x+Δx,y)+Ψ(x-Δx,y)+Ψ(x,y+Δy)+Ψ(x,y-Δy)-4Ψ(x,y)]/(Δl)^2 = 0*  \\
Por lo tanto, en su forma discreta, la Ecuación de Laplace en 2D es equivalente a la siguiente igualdad: \\
*Ψ(x,y) = (1/4)[Ψ(x+Δx,y)+Ψ(x-Δx,y)+Ψ(x,y+Δy)+Ψ(x,y-Δy)] = Ψ(x,y)_(local)*


**Ejemplo:** Resolver la Ecuación de Laplace dentro de un cuadrado de lados L, donde las condiciones de frontera son: *Ψ(x,0) = Ψ(L,y) = Ψ(0,y) = 100*, *Ψ(x,L) = 0*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_len = 100
y_len = 100
Psi_sol = np.zeros([x_len,y_len])
```

```python
# Condiciones de frontera de la ecuación:

x0 = 100
xL = 100
y0 = 100
yL = 0

Psi_sol[:,0].fill(x0) # Condición cuando x=0
Psi_sol[:,-1].fill(xL) # Condición cuando x=L
Psi_sol[-1].fill(y0) # Condición cuando y=0
Psi_sol[0].fill(yL) # Condición cuando y=L
```

```python
plt.imshow(Psi_sol)
plt.colorbar()
plt.show()
```

```python
# Solución de la Ecuación de Laplace por método de relajación:

for iterations in range(1000):
  for i in range(1,x_len-1):
    for j in range(1,y_len-1):
      Psi_sol[i,j] = (Psi_sol[i+1,j] + Psi_sol[i-1,j] + Psi_sol[i,j+1] + Psi_sol[i,j-1])/4
```

```python
plt.imshow(Psi_sol)
plt.colorbar()
plt.axis('off')
plt.show()
```

**Ejemplo:** Resolver la Ecuación de Laplace dentro de un cuadrado de lados L, donde las condiciones de frontera son: *Ψ(0,y) = Ψ(L,y) = sin(y)*, *Ψ(x,0) = Ψ(x,L) = sin(x)*

```python
X = np.linspace(0,2*np.pi,x_len)
plt.plot(X,np.sin(X),'--m')
plt.title('Condiciones de frontera',size=18)
plt.xlabel('$x$',size=12)
plt.ylabel(r'$\Psi(x)$',size=12)
plt.show()
```

```python
# Condiciones de frontera de la ecuación:

Psi_sol = np.zeros([x_len,y_len])
Psi_sol[:,0] = 100*np.sin(X) # Condición cuando x=0
Psi_sol[:,-1] = 100*np.sin(-X) # Condición cuando x=L
Psi_sol[-1] = 100*np.sin(X) # Condición cuando y=0
Psi_sol[0] = 100*np.sin(-X) # Condición cuando y=L

plt.imshow(Psi_sol)
plt.colorbar()
plt.axis('off')
plt.show()
```

```python
# Solución de la Ecuación de Laplace por método de relajación:

for iterations in range(1000):
  for i in range(1,x_len-1):
    for j in range(1,y_len-1):
      Psi_sol[i,j] = (Psi_sol[i+1,j] + Psi_sol[i-1,j] + Psi_sol[i,j+1] + Psi_sol[i,j-1])/4

plt.imshow(Psi_sol)
plt.colorbar()
plt.axis('off')
plt.show()
```
