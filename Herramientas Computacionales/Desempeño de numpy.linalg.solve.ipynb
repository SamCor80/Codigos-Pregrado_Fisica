```python
"""
Samuel Cortes Salazar
"""

import numpy as np
import time
import matplotlib.pyplot as plt



N = [10,100,200,500,1000,1200,1500,2000,2500,3000,4000,5000]

tiempo = []
desv_estandar = []
for n in N:
  t = []
  for i in range(0,10):
    A = np.random.rand(n,n)
    b = np.random.rand(n)
    inicio_t = time.process_time_ns() # Inicio del contador de tiempo
    x = np.linalg.solve(A,b)
    t.append(time.process_time_ns() - inicio_t) # Fin del contador de tiempo
  tiempo.append(np.mean(t))
  desv_estandar.append(np.std(t))

#print(tiempo)
#print(desv_estandar)

plt.errorbar(N, [t/tiempo[0] for t in tiempo], yerr = [3*d/tiempo[0] for d in desv_estandar], fmt = '*') # Normalizamos tanto al tiempo como la desv. estándar, al dividir por 'tiempo[0]'
#plt.xscale('symlog')
#plt.yscale('symlog')
plt.xlabel('Tamaño de la matriz cuadrada')
plt.ylabel('Tiempo de cómputo')
plt.title('Eficiencia de la función np.linalg.solve')
plt.grid()
plt.savefig('desempeño_tiempos.pdf')
```
