# Facebook Network

```python
Edges=np.loadtxt('facebook_combined.txt')   # Archivo importado, se encuentra en Classroom
```

```python
G=nx.from_edgelist(Edges)
```

```python
matA=nx.adjacency_matrix(G).todense()
```

```python
plt.imshow(matA)
plt.show()
```

```python
Nv=len(matA)
Degrees=[]
for i in range(Nv):
    Degrees.append(sum(matA[i]))
```

```python
plt.plot(Degrees)
plt.show()
```

```python
vals,freqs=np.unique(Degrees,return_counts=True)
```

```python
plt.loglog(vals,freqs,'k.')
plt.xlabel(r'$k$',size=20)
plt.ylabel(r'$f(k)$',size=20)
X=np.array([15.0,1000.0])
plt.loglog(X,2500*X**(-1.2),'r--')
plt.show()
```

```python
freqs=freqs/sum(freqs)
plt.loglog(vals,freqs,'k.')
plt.xlabel(r'$k$',size=20)
plt.ylabel(r'$p(k)$',size=20)
X=np.array([15.0,500.0])
plt.loglog(X,0.7*X**(-1.2),'r--')
plt.show()
```

```python
Degrees=np.array(Degrees)
```

```python
Degree_nn=[]
for i in range(Nv):
    Degree_nn.append(sum(matA[i]*Degrees)/Degrees[i])
```

```python
plt.loglog(Degrees,Degree_nn,'k.',alpha=0.1)
plt.loglog([1,1000],[1,1000],'r--')
plt.ylabel(r'$k^{(nn)}_i$',size=20)
plt.xlabel(r'$k_i$',size=20)
plt.show()
```
